```python
# importing required modules 
import PyPDF2 
```

```python
# creating a pdf file object 
pdfFileObj = open('/content/drive/MyDrive/DATA SCIENCE/CASE STUDY 2/Contractions.pdf', 'rb') 
# creating a pdf reader object 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
    
# printing number of pages in pdf file 
print(pdfReader.numPages) 
    
# creating a page object 
pageObj = pdfReader.getPage(0) 
    
# extracting text from page 
text = pageObj.extractText()
    
# closing the pdf file object 
pdfFileObj.close() 
```

```python
text
```

```python
contractions_list = []
for each in text.split():
  if "’" in each:
    # print(each)
    contractions_list.append(each)
```

```python
with open('/content/drive/MyDrive/DATA SCIENCE/CASE STUDY 2/Contractions.pkl','wb') as f:
  pickle.dump(contractions_list,f)
```

```python
homophones_list = []
```

```python
# creating a pdf file object 
pdfFileObj = open('/content/drive/MyDrive/DATA SCIENCE/CASE STUDY 2/The-Big-List-of-Homophones.pdf', 'rb') 
# creating a pdf reader object 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
    
# printing number of pages in pdf file 
pg = pdfReader.numPages

for i in range(pg):
  # creating a page object 
  pageObj = pdfReader.getPage(i) 
    
  # extracting text from page 
  text = pageObj.extractText()

  text = text.split("\n")

  for each in text:
    if "/" in each:
      homophones_list.append(each)


    
# closing the pdf file object 
pdfFileObj.close() 
```

```python
homophones_list
```

```python
with open('/content/drive/MyDrive/DATA SCIENCE/CASE STUDY 2/Homophones_list.pkl','wb') as f:
  pickle.dump(homophones_list,f)
```
